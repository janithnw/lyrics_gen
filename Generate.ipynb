```python
from torch.utils.data import Dataset, random_split
from transformers import GPT2Tokenizer, TrainingArguments, Trainer, GPT2LMHeadModel
import pickle
import torch
from tqdm.auto import tqdm
from IPython.display import Markdown, display
```

```python
def printmd(string):
    display(Markdown(string))
```

```python
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
```

```python
tokenizer = GPT2Tokenizer.from_pretrained('gpt2-medium', bos_token='<|startoftext|>',
                                          eos_token='<|endoftext|>', pad_token='<|pad|>')
```

```python
model = GPT2LMHeadModel.from_pretrained('/scratch/jnw301/lyrics_results/checkpoint-10000').to(device)
```

```python
prompt = "Boys round here, Lyrics[Verse 1]"
generated = tokenizer(prompt + "\n<|startoftext|> ", return_tensors="pt").to(device)
sample_outputs = model.generate(**generated, do_sample=True, top_k=50, 
                                max_length=1000, top_p=0.95, temperature=1.5, num_return_sequences=20, bos_token='<|startoftext|>',
                                          eos_token='<|endoftext|>', pad_token='<|pad|>')

 
```

```python
for i, sample_output in enumerate(sample_outputs):
    output = tokenizer.decode(sample_output, skip_special_tokens=True).replace(prompt, '')
    printmd("**{}:\nPrompt: {}**\n".format(i, prompt))
    print(output)
    print('\n\n\n~~~~~~~~~~~~~~~')
```
